```python
import pickle
from matplotlib import pyplot as plt
import numpy as np
```

```python
with open('../stats_SGD.pkl','rb') as f:
    l_loss, c_loss, itr = pickle.load(f)
```

```python
l_loss = np.asarray(l_loss)
c_loss = np.asarray(c_loss)
itr = np.asarray(itr)
```

```python
plt.plot(itr,l_loss+c_loss)
plt.xlabel('Iterations')
plt.ylabel('Loss')
```
