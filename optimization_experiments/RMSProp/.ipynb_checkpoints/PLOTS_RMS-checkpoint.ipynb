```python
import pickle
from matplotlib import pyplot as plt
import numpy as np
```

```python
with open('../pickles/stats_RMSProp.pkl','rb') as f:
    l_loss, c_loss, itr = pickle.load(f)
```

```python
l_loss = np.asarray(l_loss)
c_loss = np.asarray(c_loss)
itr = np.asarray(itr)
```

```python
fig = plt.figure(figsize=(15, 5)) # Create matplotlib figure
ax = fig.add_subplot(111) # Create matplotlib axes

ax.set_xlabel('Iterations', fontsize=14)
ax.set_ylabel('Loss', fontsize=14)

ax.set_title('RMSProp Optimizer Training Loss', fontsize=15)
ln_rms = ax.plot(itr[10:700],l_loss[10:700]+c_loss[10:700], color='blue', label='RMSProp Total Loss')
ln_Crms = ax.plot(itr[10:700],c_loss[10:700], color='red', label='RMSProp Confidence Loss')
ln_Lrms = ax.plot(itr[10:700],l_loss[10:700], color='green', label='RMSProp Location Loss')

lns = ln_rms + ln_Crms + ln_Lrms
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc=0)
ax.legend()
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.savefig('rms_optimizers_training_loss_separate.jpeg', bbox_inches='tight')
```
