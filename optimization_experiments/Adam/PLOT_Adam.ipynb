```python
import pickle
from matplotlib import pyplot as plt
import numpy as np
```

```python
with open('../pickles/stats_Adam.pkl','rb') as f:
    adam_Lloss, adam_Closs, adam_itr = pickle.load(f)
```

```python
adam_Lloss = np.asarray(adam_Lloss)
adam_Closs = np.asarray(adam_Closs)
adam_itr = np.asarray(adam_itr)
```

```python
with open('../pickles/stats_RMSProp.pkl','rb') as f:
    rms_Lloss, rms_Closs, rms_itr = pickle.load(f)
```

```python
rms_Lloss = np.asarray(rms_Lloss)
rms_Closs = np.asarray(rms_Closs)
rms_itr = np.asarray(rms_itr)
```

```python
with open('../pickles/stats_SGD.pkl','rb') as f:
    sgd_Lloss, sgd_Closs, sgd_itr = pickle.load(f)
```

```python
sgd_Lloss = np.asarray(sgd_Lloss)
sgd_Closs = np.asarray(sgd_Closs)
sgd_itr = np.asarray(sgd_itr)
```

```python
fig = plt.figure(figsize=(15, 5)) # Create matplotlib figure
ax = fig.add_subplot(111) # Create matplotlib axes
# ax2 = ax.twinx()
# ax3 = ax.twinx()

ax.set_xlabel('Iterations', fontsize=14)
ax.set_ylabel('Loss', fontsize=14)

ax.set_title('ADAM Optimizers Training Loss', fontsize=15)
ln_adam = ax.plot(adam_itr[7:700],adam_Lloss[7:700]+adam_Closs[7:700], color='blue', label='ADAM Total Loss')
ln_Cadam = ax.plot(adam_itr[7:700],adam_Closs[7:700], color='red', label='ADAM Confidence Loss')
ln_Ladam = ax.plot(adam_itr[7:700],adam_Lloss[7:700], color='green', label='ADAM Location Loss')

lns = ln_adam + ln_Cadam + ln_Ladam
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc=0, prop={'size': 12})
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.savefig('adam_optimizers_training_loss_separate.jpeg', bbox_inches='tight')
```
