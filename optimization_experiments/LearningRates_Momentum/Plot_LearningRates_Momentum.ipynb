```python
import pickle
from matplotlib import pyplot as plt
import numpy as np
```

### LR 1e^-1

```python
with open('../pickles/stats_lr_+0.1.pkl','rb') as f:
    l_loss, c_loss, itr = pickle.load(f)
    
l_loss = np.asarray(l_loss)
c_loss = np.asarray(c_loss)
itr = np.asarray(itr)

fig = plt.figure(figsize=(15, 5)) # Create matplotlib figure
ax = fig.add_subplot(111) # Create matplotlib axes

ax.set_xlabel('Iterations', fontsize=14)
ax.set_ylabel('Loss', fontsize=14)

ax.set_title('Learning Rate 1e-1 Training Loss', fontsize=15)
ln_sgd = ax.plot(itr,l_loss+c_loss, color='blue', label='LR 1e-1 Total Loss')
ln_Csgd = ax.plot(itr,c_loss, color='red', label='LR 1e-1 Confidence Loss')
ln_Lsgd = ax.plot(itr,l_loss, color='green', label='LR 1e-1 Location Loss')

lns = ln_sgd + ln_Csgd + ln_Lsgd
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc=0)
ax.legend()
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.savefig('lr_e_neg1_training_loss_separate.jpeg', bbox_inches='tight')
```
